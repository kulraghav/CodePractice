Visualizing basic statistical distributions
https://seaborn.pydata.org/tutorial/distributions.html

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
sns.set(color_codes=True)
%matplotlib inline
```

```python
np.random.seed(42) # fixes the random seed

x = np.random.normal(size=1000)
sns.distplot(x)
plt.show()
```

```python
iris = sns.load_dataset("iris")
print ("type(iris): {} iris.shape: {}".format(type(iris), iris.shape))
sns.pairplot(iris)
plt.show()
```
